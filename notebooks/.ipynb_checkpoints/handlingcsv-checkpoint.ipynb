```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
if __name__ == "__main__":
    filename = '../data/lisbon.csv'
    #filename = '../data/oxford.csv'
    
       
    data = np.genfromtxt(filename, delimiter=',', names=True)
    hour = data['hour']
    precipitation = data['precipitation']
    temperature = data['temperature']
    wind = data['wind']
```

```python
hour
```

```python
plt.plot(temperature)
plt.plot(wind)
plt.plot(precipitation)
plt.show()
```

```python
plt.plot(hour,temperature)
plt.plot(hour,wind)
plt.plot(hour,precipitation)
plt.show()
```

```python
f, axarray = plt.subplots(3, 1, sharex=True, figsize=(7,6))

ax = axarray[0]
ax.plot(temperature)
ax.set_ylabel('Temperature [C]')


ax = axarray[1]
ax.plot(wind)
ax.set_ylabel('Wind speed [km/s]')

ax = axarray[2]
ax.plot(precipitation)
ax.set_ylabel('Precipitation [mm]')

plt.show()
```

```python
hour = hour.reshape(24,4)
temperature = temperature.reshape(24,4)
wind = wind.reshape(24,4)
precipitation = precipitation.reshape(24,4)

plt.plot(hour,temperature)
plt.plot(hour,wind)
plt.plot(hour,precipitation)
plt.show()
```
