```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
if __name__ == "__main__":
    filename = '../data/temperatures.csv'

    data = np.genfromtxt(filename, delimiter=',')
    
```

```python
data
plt.plot(data)
plt.legend(['Lisbon','London','Paris'])

plt.show()
```

```python
lisbon = data[:,0]
london = data[:,1]
paris = data[:,2]
```

```python
plt.plot(lisbon, label='Lisbon')
plt.plot(london, label='London')
plt.plot(paris, label='Paris')
plt.xlabel('Hour')
plt.ylabel('Temperature [C]')
plt.legend()
plt.show()
```

```python
f, axarray = plt.subplots(3, 1, sharex=True, sharey=True, figsize=(7,6))

ax = axarray[0]
ax.plot(lisbon, label='Lisbon')

ax = axarray[1]
ax.plot(london, label='London')

ax.set_ylabel('Temperature [C]')

ax = axarray[2]
ax.plot(paris, label='Paris')

ax.set_ylabel('Hour')

plt.show()
```

```python
plt.subplot(311)
plt.plot(lisbon, label='Lisbon')
plt.subplot(312)
plt.plot(london, label='London')
plt.subplot(313)
plt.plot(paris, label='Paris')
plt.legend()
plt.show()
```

```python
# Now we can do a function that does this reshaping for us
def hours2days(data):
    days = len(data)/24
    data = data.reshape(days,24)
    return data
```
